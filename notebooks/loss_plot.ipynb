```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
def plot_loss(filename, ylim=None):
    data = pd.read_csv(filename, header=None)
    plt.plot(data)
    if ylim is not None:
        plt.ylim(ylim)
    plt.yscale("log")
    plt.show()
```
